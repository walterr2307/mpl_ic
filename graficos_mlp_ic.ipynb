```python
import matplotlib.pyplot as plt
import re
import os
```

```python
with open("/content/MLP - 3 CAMADAS OCULTAS.txt", "r", encoding="utf-8") as arquivo:
    linhas = arquivo.read().splitlines()
```

```python
def definirNovaLinha(linha):
  if len(linha) > 0:
    if linha[0] == 'G':
      return linha.replace(' ', '')

  return None

def definirNovasLinhas(linhas):
  texto = ''

  for linha in linhas:
    nova_linha = definirNovaLinha(linha)

    if nova_linha is not None:
      texto += nova_linha + '\n'

  return texto.split('\n')

def separarMetricas(linhas):
  loss = []
  tempo = []
  media = []

  for linha in linhas:
    partes = re.split(r'[:|]', linha)

    if len(partes) > 13:
      partes[11] = partes[11][:-1]
      partes[13] = partes[13][:-1]

      loss.append(float(partes[9]))
      tempo.append(float(partes[11]))
      media.append(float(partes[13]))

  return [loss, tempo, media]

def separarPorGerecao(num_individuos, num_geracoes, metricas):
  conj_loss = []
  tempos = []
  medias = []

  for _ in range(num_geracoes):
    conj_loss.append([])
    tempos.append([])
    medias.append([])

    for _ in range(num_individuos):
      conj_loss[-1].append(metricas[0][0])
      tempos[-1].append(metricas[1][0])
      medias[-1].append(metricas[2][0])

      for i in range(3):
        del metricas[i][0]

  return conj_loss, tempos, medias
```

```python
linhas = definirNovasLinhas(linhas)
metricas = separarMetricas(linhas)

conj_loss, tempos, medias = separarPorGerecao(6, 5, metricas)
```

```python
def plotar_por_geracao(dados, titulo, ylabel):
    plt.figure()

    for i, geracao in enumerate(dados):
        plt.plot(
            range(1, len(geracao) + 1),
            geracao,
            marker='o',
            label=f"Geração {i+1}"
        )

    plt.xlabel("Indivíduos")
    plt.ylabel(ylabel)
    plt.title(titulo)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plotar_por_geracao(
    conj_loss,
    "Loss por Indivíduo em Cada Geração",
    "Loss"
)
```

```python
plotar_por_geracao(
    tempos,
    "Tempo de Treinamento por Indivíduo em Cada Geração",
    "Tempo (s)"
)
```

```python
plotar_por_geracao(
    medias,
    "Fitness por Indivíduo em Cada Geração",
    "Fitness (%)"
)
```
